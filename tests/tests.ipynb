# 1. Setup

```python
import httpx

BASE_URL = "http://localhost:8000"
client = httpx.Client(base_url=BASE_URL)
```

# 2. Customers test
## 2.1 get by filter

```python
params = {
    "country": "USA",
    "state" : "NY"
}

response = client.get("/customers", params=params)

print(response.status_code)
print(response.json())
```

# 2.2 By Id

```python
customer_id = 103

response = client.get(f"/customers/{customer_id}")

print(response.status_code)
print(response.json())
```

## 2.3 POST customer

```python
new_customer = {
    "customerNumber": 10000003,
    "customerName": "Test Name",
    "contactLastName": "Last",
    "contactFirstName": "First",
    "phone": "123456789",
    "addressLine1": "Test Street",
    "city": "NYC",
    "country": "USA",
}

response = client.post("/customers", json=new_customer)

print(response.status_code)
print(response.json())
```

```python
response = client.get(f"/customers/{10000003}")

print(response.status_code)
print(response.json())
```

## 2.4 PUT customer

```python
update_data = {
    "customerNumber": 10000003,
    "customerName": "Updated Name",
    "contactLastName": "UpdatedLast",
    "contactFirstName": "UpdatedFirst",
    "phone": "987654321",
    "addressLine1": "Updated Street",
    "city": "NYC",
    "country": "USA",
}

response = client.put("/customers/10000003", json=update_data)

print(response.status_code)
print(response.json())
```

## 2.5 Delete customer

```python
response = client.delete("/customers/10000003")

print(response.status_code)
print(response.json())
```

```python
response = client.delete("/customers/10000003")

print(response.status_code)
print(response.json())
```
